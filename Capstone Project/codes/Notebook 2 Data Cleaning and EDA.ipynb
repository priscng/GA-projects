### Preprocessing text

- Text preprocessing will be done for review_header, pro, and con columns.
- The text preprocessing includes:
    - Removing the punctuations and stopwords.
    - Tokenize the text.
    - Lemmatize the text.

```python
# Create a function to define a mapping between wordnet tags and the words
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
```

**Note:** Lemmatization by default, treats all the words passed to it as nouns. Therefore, we should lemmatize the words based on their respective parts of speech by using the function above.

```python
# Create a function to preprocess the text
def preprocess_text(dataset, col, new_col):
    
    text_processed = []
    for text in dataset[col]:

        # Remove punctuation and convert to lowercase
        text = "".join([word.lower() for word in text if word not in string.punctuation])
        
        # Tokenize text
        token = RegexpTokenizer(r'\s+', gaps=True)
        text_split = token.tokenize(text.lower())

        # Remove stopwords
        text_split = [word for word in text_split if word not in (stopwords)]
        
        # Initialize the lemmatizer and lemmatize the text
        wn = WordNetLemmatizer()
        text_split = [wn.lemmatize(word, get_wordnet_pos(word)) for word in text_split]
        
        text_processed.append(text_split)
    
    # Create new column for the cleaned text
    dataset[new_col] = text_processed
    dataset.loc[:, new_col] = dataset[new_col].apply(lambda x: ' '.join(map(str, x)))
```

### Word cloud visualization

Word Cloud or Tag Cloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency or importance. We will plot a bigram word cloud for this visualization.

```python
# Create a function to generate and plot bigram word cloud
def create_wordcloud(dataset, column, filename=None):
    text = ' '
  
    # iterate through column 
    for word in dataset[column]: 

        # Convert each word to string 
        word = str(word) 

        # Split the value 
        values = word.split() 

        # Converts into lowercase 
        for i in range(len(values)): 
            values[i] = values[i].lower() 

        for words in values: 
            text = text + words + ' '
    
    # Creating wordcloud with text as argument in the .generate() method
    wordcloud = WordCloud(max_words=60, width=744, height=544, min_word_length=3, 
                           collocation_threshold=3, background_color='white').generate(text)

    # Display the generated word cloud
    plt.figure(figsize=(14,8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    
    # Save image if filename is specified
    if filename != None:
        wordcloud.to_file(f'../images/{filename}.png')
```

### N-gram visualization with Count Vectorizer

- CountVectorizer counts the number of word occurrences and helps identify the common text patterns in the documents.
- CountVectorizer is used to perform trigram and quadgram analysis on the pros, cons, and headers reviews to help us understand the employees' views on working in a four-day work week company.

```python
# Create a function to instantiate, fit and transform using CountVectorizer with ngram
def countVec(dataset, n1, n2, column, ngram, title, filename=None):
    cvec = CountVectorizer(ngram_range = (n1, n2))
    cvec.fit(dataset[column])
    cvec_transformed = cvec.transform(dataset[column])

    df_transformed = pd.DataFrame(cvec_transformed.todense(),
                                columns = cvec.get_feature_names())
    
    # Plot the bar graph
    plt.figure(figsize=(20, 10))
    df_transformed.sum().sort_values(ascending = True)[-20:].plot(kind='barh')
    plt.title(title, fontsize=16, fontweight='bold')
    plt.ylabel('Words', fontsize=12)
    plt.xlabel('Frequency', fontsize=12)

    # Count the words frequency, convert it to a dataframe and display it
    sum_words = cvec_transformed.sum(axis = 0) 
    words_freq = [(word, sum_words[0, i]) for word, i in cvec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)
    words_freq_df = pd.DataFrame(words_freq)
    words_freq_df.columns=[ngram, 'Frequency']
    display(words_freq_df[:20])
    
    # Save image if filename is specified
    if filename != None:
        plt.savefig(f'../images/{filename}.png', transparent=True, dpi=300)
        #plt.savefig(f'../images/{filename}.png')
```

### N-gram visualization with TFIDF Vectorizer

- TfidfVectorizer is a count vectorizer that includes some consideration for the length of the document, and also how common the word is across other text messages.
- Similarly, TfidfVectorizer is used to perform trigram and quadgram analysis on the pros, cons, and headers reviews to help us understand the employees' views on working in a four-day work week company.
- Similar top words are observed with TfidfVectorizers but the frequency is lower as this vectorizer includes other considerations as mentioned above.

```python
# Create a function to instantiate, fit and transform using TFIDF Vectorizer with ngram
def tfidf(dataset, n1, n2, column, ngram, title, filename=None):
    tfidf_vect = TfidfVectorizer(ngram_range = (n1, n2))
    tfidf_vect.fit(dataset[column])
    tfidf_transformed = tfidf_vect.transform(dataset[column])

    df_transformed = pd.DataFrame(tfidf_transformed.todense(),
                                columns = tfidf_vect.get_feature_names())
    
    # Plotting bar graph
    plt.figure(figsize=(15, 10))
    df_transformed.sum().sort_values(ascending = True)[-20:].plot(kind='barh')
    plt.title(title, fontsize=16, fontweight='bold')
    plt.ylabel('Words', fontsize=12)
    plt.xlabel('Frequency', fontsize=12)
    
    # Count the words frequency, convert it to a dataframe and display it
    sum_words = tfidf_transformed.sum(axis = 0) 
    words_freq = [(word, sum_words[0, i]) for word, i in tfidf_vect.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)
    words_freq_df = pd.DataFrame(words_freq)
    words_freq_df.columns=[ngram, 'Frequency']
    display(words_freq_df[:20])
    
    # Save image if filename is specified
    if filename != None:
        plt.savefig(f'../images/{filename}.png', transparent=True, dpi=300)
```

### Sentiment analysis

- Sentiment analysis is a commonly used NLP (natural language processing) technique to determine whether the text is positive, negative, or neutral. We will analyse employee satisfaction towards a four-day work week company based on the review sentiment.
- For this analysis, we will use the Sentiment Intensity Analyser which uses the VADER Lexicon. VADER (Valence Aware and sEntiment Reasoner) is a rule-based sentiment analysis tool that calculates the text emotions and determines whether the text is positive, neutral, or, negative. This analyzer calculates text sentiment and produces four different classes of output scores: positive, negative, neutral, and compound. 

- A compound score is a metric that sum up all the lexicon ratings which have been normalized between -1 (most extreme negative) and +1 (most extreme positive). The pos, neg, neu scores represent the ratios for proportions of text that fall in each category, and the sum of positive, negative, and neutral scores gives 1.

- In this analysis, we will use the compound score as the metric.

Net sentiment score is based on the compound score. We define the sentiment is:

- positive if compound score is > 0
- negative if compound score is <= 0

```python
def vader_analysis(compound):
    if compound <= 0 :
        return 'Negative'
    else:
        return 'Positive'
```

```python
# Apply vader analysis function
company_train['header_analysis'] = company_train['header_compound'].apply(vader_analysis)
company_train['pro_analysis'] = company_train['pro_compound'].apply(vader_analysis)
company_train['con_analysis'] = company_train['con_compound'].apply(vader_analysis)
```

```python
# Create a function to plot donut piechart
def plot_donut_piechart(dataset, column, title, filename=None):
    # Count the number of pro and cons
    count = dataset[column].value_counts()
    
    # create data
    names = ['Positive', 'Negative']
    size = count

    # Create a circle at the center of the plot
    my_circle = plt.Circle((0,0), 0.7, color='white')

    # Plot donut piechart
    plt.figure(figsize=(10,5))
    plt.pie(size, autopct='%1.1f%%', labels=names, pctdistance=0.5, startangle=45)
    p = plt.gcf()
    p.gca().add_artist(my_circle)
    plt.title(title, fontsize=14, fontweight='bold')

    # Save image if filename is specified
    if filename != None:
        plt.savefig(f'../images/{filename}.png', transparent=True, dpi=300)
```

```python
# Plot sentiment category donut piechart for review header
plot_donut_piechart(company_train, 'header_analysis', 'Sentiment Analysis for Header Reviews')
```

### Employee tenure

- Employee tenure is the length of time an employee works for an employer.
- The average employee tenure is around 21 months ([source](https://www.forbes.com/sites/forbestechcouncil/2021/07/28/why-your-employees-are-leaving-en-masse-and-the-surprising-factor-that-will-keep-them/?sh=6af005fd40fb)). 
- We will categorise the employee tenure as below:

    - Less than 21 months (industry standard is 21 months and the turnover rate is considered high when an employee stays less than a year in a company)
    - From 21 to 36 months (In general, an employee learns and adapts to a new company in the first 2 years. Employee starts to contribute from the 3rd year onwards)
    - More than 36 months

```python
# Create a funtion to assign the tenure category
def tenure_cat(tenure):
    if tenure < 1.8:
        return 'Less than 21 months'
    elif tenure > 3:
        return 'More than 36 months'
    else:
        return 'From 21 to 36 months'
```

```python
# Create a column for tenure category and apply tenure_cat function
company_train['tenure_cat'] = company_train['tenure'].apply(tenure_cat)
```
